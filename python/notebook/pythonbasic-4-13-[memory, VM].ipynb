```python
'''
컴퓨터공학 5대 과목
1. 프로그래밍언어 - 파이썬, 자바, c...
2. 자료구조 - 링크드테스트, 스택/큐
3. 알고리즘 - binary search, sorting
4. 컴퓨터구조 - CPU, 레지스터
5. OS(운영체제) - 메인메모리(RAM, Hard Disc)

'''

## ComputerScienceSchool/lecture/memory $ VM
'''
c언어 register 처리 - ppt참고
프로그램 카운터는 어셈블리어의 주소값을 받아놓는다.
인스트럭션 레지스터는 해당 주소값의 어셈블리어를 실행한다.
스택포인터는 스택에 인스트럭션 레지스터가 처리한 값을 쌓거나 용량을 할당, 또 데이터를 지우는 역할을 한다.
ebp를 미리 스택으로 향하게 해놓은 다음 다음 코드를 위하여 12byte를 확보(SUB esp, 12)하여 스택 위치를 변경시킨다. 
SUB esp, 12 - esp에서 12를 빼서 esp에 넣는다. 
MOV a, 0Ah A의 값을(10)을 ebp로 놓게 한 후 stack에 다시 전역변수를 쌓는다. 
스택에 옮겨진 a(10)과 b(20)은 스택에서 연산할 수 없기 때문에 반드시 범용 레지스터로 옮겨 연산해야 한다.
ADD eax, b - a값과 b값을 eax라는 범용 레지스터로 옮겨서 연산, 레지스터에 한번 저장한다!!
MOV c, eax - 레지스터 eax에 저장되어 있는 연산값을 c로 옮긴 후 stack에 다시 저장한다.
## 스택에서는 연산을 할 수 없으므로 a, b 값을 레지스터에 옮겨 연산한 후 연산 결과또한 레지스터에 저장하여 다시 스택으로 옮긴다.
스택 포인터는 스택에 쌓아올린 데이터 바로 위를 향하게 되어있다. (그래서 사전에 12를 빼서 올린 것. )
스택 포인터는 스택프레임을 위해 미리 공간을 확보하는 역할을 하므로. 
print() 출력- 스택에 a, b, c 값은 그대로 존재한다. 
mov ebp, esp - ebp가 a(10)을 가리키고 스택 포인터가 ebp를 가리키게 되면 다음에 들어올 데이터는 값이 덮어씌워져버린다. 
스택포인터가 이전 데이터 위치를 가리키게 해버리면 스택에 쌓여있던 자료는 새로운 자료에 의해 지워지게 되는 것(os의 데이터 삭제 방법)
스택에 저장되어 있던 이전 ebp값을 레지스터의 ebp값으로 향하게 하고 스택포인터를 스택의 ebp로 향하게 하면 ebp까지 삭제할 수 있다. 

c언어 - code부터 stack까지 모두 사용가능하다. 
     - 연산은 레지스터에서 한다.
'''


'''
파이썬을 포함한 인터프리터 언어 - 전역변수를 포함하여 모든 데이터를 heap에만 저장한다. 
파이썬, java
인터프리터 언어의 모든 객체는 heap에 저장. 연산은 stack에서 함. 
인터프리터 언어의 컴파일 과정은 c언어를 따라하지만 그 위치가 다르다. ==> virtual machine
'''


'''
virtual machine
레지스터 기반 머신(register-based machine) - 연산자료, 연산결과를 레지스터에 저장
가상 스택머신(virtual stack machine) - 연산자료, 연산결과를 스택에 가지고와 연산하고 힙에 반환
operand stack - 힙에 있는 객체들을 스택에 저장하는 공간을 operand stack이라고 한다. 
                연산하려는 값들을 임시저장하는 공간.

## 바이트코드 해석
LOAD FAST - 힙에서 로드한 a 객체를 오퍼렌드 스택에 쌓음
LOAD FAST - 힙에서 로드한 b 객체를 오퍼렌드 스택에 쌓음
BINARY ADD - a + b 값은 스택에 저장되어 있음. 
STORE FAST - 스택에서 연산한 결과를 힙으로 다시 반환

LOAD FAST - 힙에서 로드한 c 객체를 오퍼렌드 스택에 들고옴
RETURN VALUE - c에 연산결과를 참조하여 프라이빗 힙으로 다시 이동. 
'''


'''
segmentation의 특징>>

"스택은 빠르고 힙은 느리다." = SMA와 DMA의 차이
스택은 스택포인터가 있어 순차적으로 쌓이므로 주어진 용량대로 할당을 하기만 하면 된다.
관련데이터가 차례대로 쌓이므로 spacial locality를 보장한다.
힙은 메타데이터와 실제 데이터정보가 쌓인다.
메타데이터 때문에 할당용량을 초과하는 용량을 차지하게 된다.
용량을 할당할 때도 메타데이터에 공간이 있는지 물어보고 순회해야하므로 힙이 느린 것.
힙은 데이터를 할당하는 시점, 지우는 시점을 마음대로 조정할 수 있으므로 사용한다. 

code와 data는 할당할 크기가 이미 정해져있다.(SMA, Static Memory Allocation)
stack공간도 프로세스당 1M로 용량이 정해져있다.(초기값 지정하여 늘릴 수 있음) 
stack이 다 차면 stack-overflow에러가 난다.
그러나 heap은 디폴트로 1M를 잡아놨지만 스택의 영역을 만날 때까지 넣을수록 늘어난다. 
따라서 heap은 동적인 메모리 할당을 지원한다. (DMA, Dynamic Memory Allocation)
정해진 code, data, stack 을 제외한 영역에서 heap은 마음대로 사용할 수 있다.(최대 2GB)
'''
```
