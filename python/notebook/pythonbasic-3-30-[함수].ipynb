```python
# arguments - 인자의 갯수를 정의하지 않을 때
# *args는 사용자에 의해 함수에 몇개의 argument가 넘겨질지 모를 때 사용
# arguments : 파라미터의 갯수를 지정하기 힘들 때 사용한다.
# arguments들을 받아서 for문에 돌려 arguments의 parameter(인자)
# 들을 하나하나 sum에 저장하는 식. 


# pass - 함수의 기능을 정의하고 딱히 실행하지 않을 때 쓴다. 

def mul_sum(*args):    # *를 변수에 붙여야 파이썬의 아규먼츠로 인식함. 
    pass               # 안 붙이면 args라는 변수로 인식
    
    sum = 0            # 합 초기화
    for i in args:
        sum += i       # 또는 sum = sum += i
    return sum
```

```python
mul_sum(1, 2, 3, 4, 5,)
```

```python
def mul_sum(*phone):
    pass
    sum = 0
    for i in phone:
        sum += i
    return sum
```

```python
# 정보 크롤링할 때 쓰는 핵심 함수
# 사이트 도메인별로 쿼리를 다르게 지정해서 사용가능하다.
def kwargs_url(server, port, **query):
    url = "https://" + server + ":" + port + "?"
    for key in query.keys():
        url += key + "=" + query[key] + "&"
    return url
```

```python
kwargs_url("www.naver.com", "8080", query="고양이", where="nexearch", ie="utf-8", keyboard="naver")
```

```python
'''
global이라는 명령을 사용하여 전역변수(모든 프로그램 코드에 영향)로
사용하게 되면 함수는 독립성을 잃게 된다.(외부변수에 의존적이게 된다.)
# 사용하게 되면 해당 함수와 외부에 있는 변수지정식이 묶여버려서 
# 해당 함수를 호출할 때마다 파일이 실행된다. 당연히 속도는 느려진다!!
# 다른 곳에서 변수값을 계속 가져와야 하므로!!

# 간편하다고 갖다 쓰면 함수로써의 정의를 잃어버리게 된다. 
# 쓰긴 쓰지만 부득이한 경우에만 쓴다. 
# 파이썬의 pythonic함을 극대화하기 위해서는 코드 천줄을 줄일 수 없는 이상
# 종속 변수인 global 을 사용하지 않는 것이 좋다. 

# 따라서 함수가 외부변수에 의존적이게 되는데, global 변수를 사용하면
# 변수의 정의가 프로그램 전체에 적용되고 함수가 외부에서 변수값을
# 자꾸 가져와야한다.쓰지 않는 것이 좋음. 
'''

dog = "hello"
def glob_hello():
    global dog
    dog +="world"
    return dog
```

```python
glob_hello()
```

```python
print(dog)
```

```python
# leap year
# 상단에 num = %d 칠 필요 없음.

num = eval(input("type the year: "))

for i in range(1, num + 1):
    if i % 4 == 0 and (i % 400 == 0 or i % 100 != 0):
        print("%d is a leap year" % (num))
    else:
        print("%d is not a leap year" % (num))
```

```python
def is_leap(year, leap = False):
    if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
        leap = True
        return leap

leap = is_leap(year)
    
def print_result():
    if leap == True:
        print("This is a leap year.")
    else:
        print("This is not a leap year.")
```

```python
print_result()
```
