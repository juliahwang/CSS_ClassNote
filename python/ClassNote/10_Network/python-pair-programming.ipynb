```python
# 1. ethiopian

def multifly(a):
    return a * 2

def divide(a):
    return a // 2

x = int(input("숫자를 입력하세요> "))
y = int(input("숫자를 입력하세요> "))

keeps = 0


while x != 0:
    print(x, " ", y)
    if x % 2 == 1:
        keeps += y
        
    x = divide(x)
    y = multifly(y)
       
        
print(keeps)   
```

```python
# crawling

from bs4 import BeautifulSoup
import urllib
url = "https://stackoverflow.com/questions/tagged/python"
html = urllib.request.urlopen(url)
source = html.read()
html.close()


soup = BeautifulSoup(source, "html.parser")
print(soup)

table = soup.find(id = "questions")
print(table)

all_tr = table.find_all("question-summary")
print(all_tr)
```
