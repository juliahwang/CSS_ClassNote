```python
# beautiful soup을 이용하여 웹스크랩하기
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = "https://www.rottentomatoes.com"
html = urllib.request.urlopen(url)
source = html.read()
html.close()
```

```python
soup = BeautifulSoup(source, "html.parser")
table = soup.find(id="Top-Box-Office")
print(table)
```

```python
# 추출한 값이 여러개인 경우에는 리스트에 저장된다. - for문으로 찾아내기 쉽다
# 추출한 결과에서 조건을 만족하는 모든 결과값을 all_tr에 저장.
all_tr = table.find_all("tr")
print(all_tr)
```

```python
# html 파싱하기(score, link, mv_title, amount)
for tr in all_tr:
    all_td = tr.find_all("td")
    score = all_td[0].find("span", attrs={"class": "tMeterScore"}).text
    # all_td에서 0번째 텍스트열 중에 span 태그 안의 어트리뷰트 중 class가 tMeterScore인 줄의 text를 가져온다.
    link = url + all_td[1].a.get("href")     # 데이터 클렌징 : url보이게하기
    mv_title = all_td[1].a.text
    amount = all_td[2].a.text
    amount = amount.split()[0]    # 데이터 클렌징 : 띄어쓰기 없애기
    print(score,"\n",link,"\n",mv_title,"\n",amount,"\n")    # 데이터 클렌징 : 줄바꿈
```
