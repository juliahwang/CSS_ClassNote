```python
from stack import Stack

class Calculator:
    def __init__(self, org_exp):
        self.org_exp = org_exp.replace(' ', '')
        # 수식 입력시 발생하는 모든 공백을 없애준다.
        self.postfix_exp = None
        
    # operater가 나올 때 가중치를 주는 함수
    def get_weight(self, oprt):
        if oprt == '*' or oprt == '/':
            return 9
        elif oprt == '+' or oprt == '-':
            return 7
        elif oprt == '(':
            return 5
        
    
    # 후위표기법으로 바꿔주는 함수
    # 1. 수식리스트 필요 - exp_list=[]
    # 2. 연산자를 담을 스택 필요 - oprt_stack
    # 3. for문 - ch가 수식에 있는동안 반복한다.   
    # 4.     if문 : 숫자는 무조건 수식리스트에 담기
    #              연산자는 oprt_stack에 넣기
    #                if문 : '('가 나오거나 연산자 스택이 비어있다면 연산자 스택에 넣기
    #                      if문: ')'가 나오면 이전에 넣은 '('이 나올때까지 .pop()실행. - 수식리스트에 넣는다.
    #                      else: ch의 가중치 > 스택 peek -> 스택에 쌓는다. 
    #                      ch 가중치 <= stack.peek() -> 수식리스트에 넣고 stack에 넣어준다. 
    #                      만약, oprt_stack에 연산자가 남아있다면 모두 수식리스트에 집어넣는다. ''.join(exp_list) 

    def convert_to_postfix(self):
        exp_list = []
        # self.exp_list = [] 라고 쓰면 계속 클래스 내에서 사용 가능.
        oprt_stack = Stack()
        
        for ch in self.org_exp:
            if ch.isdigit():
                exp_list.append(ch)
            elif ch == '(' or not oprt_stack:
                oprt_stack.push(ch)
            elif ch == ')':
                #for ch in oprt_stack:
                #    exp_list.append(oprt_stack.pop())
                op = oprt_stack.pop()
                while op != '(':
                    exp_list.append(op)
                    op= oprt_stack.pop()
                            
            elif self.get_weight(ch) > self.get_weight(oprt_stack.peek()):
                oprt_stack.push(ch)
                #elif self.get_weight(ch) <= self.get_weight(oprt_stack.peek()):
                #    exp_list.append(ch)
                #else:
                #    while oprt_stack != 0:
                #        exp_list.append(oprt_stack.pop())
                        
            else:
                while oprt_stack and self.get_weight(ch) <= self.get_weight(oprt_stack.peek()):
                    exp_list.append(oprt_stack.pop())
                oprt_stack.push(ch)

        while oprt_stack:
            exp_list.append(oprt_stack.pop())
        
        # 저장하기
        self.postfix_exp = ''.join(exp_list)
            
            
            
    # 연산자 구동 방식을 알려주는 함수 
    def calc_two_oprd(self, oprd1, oprd2, oprt):
        if oprt == '+':
            return oprd1 + oprd2
        elif oprt == '-':
            return oprd1 - oprd2
        elif oprt == '*':
            return oprd1 * oprd2
        elif oprt == '/':
            return oprd1 // oprd2
        
        
    # calculate 함수 
    def calculate(self):
    # 1. oprd_stack - 피연산자를 담을 스택 마련
        oprd_stack = Stack()
        
    # 2. postfix_exp 순회(for)
        for ch in self.postfix_exp:
    #    1) isdigit() -> oprd_stack에 담기
            if ch.isdigit():
                oprd_stack.push(int(ch))
    #    2) 연산자 +,-,*,/ -> stack에서 2개 빼서
            else:
                oprd2 = oprd_stack.pop()
                oprd1 = oprd_stack.pop()
                
    #    3) 연산 값 리턴하여 다시 스택에 담고
                oprd_stack.push(self.calc_two_oprd(oprd1, oprd2, ch))
    # 3. 스택에서 남은 값을 빼서 반환
        return oprd_stack.pop()
            
                
    
if __name__ == "__main__":
    calc = Calculator(a)
    calc.convert_to_postfix()
    print(calc.postfix_exp)
    print(calc.calculate())
```
